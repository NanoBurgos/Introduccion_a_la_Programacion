El factorial de un número (n!) es el resultado de multiplicar ese número por todos los anteriores hasta llegar al 1.
Por ejemplo:
• 4! = 4 × 3 × 2 × 1 = 24
• 5! = 5 × 4 × 3 × 2 × 1 = 120
