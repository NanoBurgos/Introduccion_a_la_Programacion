# **Estructura Secuencial**

La estructura secuencial es aquella en la que
una acción (instrucción) sigue a otra en
secuencia.


Las tareas ocurren de tal manera que la
salida de una es la entrada de la siguiente y
así sucesivamente hasta el final del proceso.


La estructura secuencial tiene una entrada y
una salida.

**Ejercicio 2:** Diseñe un algoritmo
que determine el
porcentaje de varones
y de mujeres que hay
en un salón de clases.

    Algoritmo
      
      //Definimos los tipos de variables
      Definir varones, mujeres, total como Entero
      Definir porcentaje_varones, porcentaje_mujeres como Real

      //Pedimos los datos
      Escribir "Ingrese la cantidad de varones en la clase: "
      Leer varones
      Escribir "Ingrese la cantidad de mujeres en la clase: "
      Leer mujeres

      //Realizamos los calculos
      total = varones + mujeres

      porcentaje_varones = (varones*100.0)/total
      porcentaje_mujeres = (mujeres*100.0)/total

      //Mostramos los resultados
      Escribir "El porcentaje de varones es: ", porcentaje_varones
      Escribir "El porcentaje de mujeres es: ", porcentaje_mujeres

    FinAlgoritmo





**Ejercicio 3:** Diseñe un algoritmo para
convertir una longitud dada en
metros a sus equivalentes en
centímetros, pies, pulgadas y
yardas.

  **Considere que:**
    
    1 metro = 100 centímetros,

    1 pulgada = 2.54 centímetros,

    1 pie = 12 pulgadas,

    1 yarda = 3 pies.




    Algoritmo
    Definir metros, centimetros, pulgadas, pies, yardas como Real

    //Pedimos al usuario que ingrese el valor
    Escribir "Ingrese la cantidad (en metros), que desea convertir en centimetros, pulgadas, pies y yardas: "
    Leer metros

    //Realizamos los calculos

    centimetros = metros*100

    pulgadas = centimetros/2.54

    pies = centimetros/(2.54*12)

    yardas = centimetros/(2.54*12*3)

    //Mostramos los calculos

    Escribir "Ingreso: " metros " , su conversion a los demas sistemas es: "
    Escribir "En centimetros: " centimetros " cm"
    Escribir "En pulgadas: " pulgadas " inches"
    Escribir "En pies: " pies " feet"
    Escribir "En yardas: " yardas " yard"
   
    FinAlgoritmo

**Ejercicio 4:** Diseñe un
algoritmo que
determine la cifra
de las unidades de
un número natural.

    Algoritmo
    
    Definir numero, unidades como Entero

    //Pedimos los datos al usuario
    Escribir "Ingrese un numero: "
    Leer numero

    //Realizamos los calculos
    unidades = numero % 10     // El resto de una division conseguimos con % o MOD

    Escribir "El digito de las unidades es: ", unidades
