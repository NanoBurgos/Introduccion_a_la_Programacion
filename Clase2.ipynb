# **Estructura Selectivas o Condicionales**

**Estructuras Selectivas**

Se ejecuta una instrucción u otra según
se cumpla o no una determinada
condición.

Pueden ser: Simples, Dobles, Multiples

# **Estructuras Selectivas SIMPLES**
1 Se evalúa una condición.

2 Si la condición se cumple se ejecutan determinadas acciones.

3 Si la condición no se cumple, no se ejecutan esas acciones.


# **1 SIMPLES**

Se evalúa la condición y si ésta da como resultado verdadero se ejecuta una determinada instrucción o grupo de instrucciones, en caso contrario se saltan dicho grupo de instrucciones.


# **2 DOBLES**

Cuando el resultado de evaluar la condición es verdadera se ejecutará una determinada instrucción o grupo de instrucciones y si el resultado es falso instrucción o grupo de instrucciones diferentes.


# **Estructuras Selectivas DOBLES**

1 Se evalúa una condición.

2 Si la condición se cumple se ejecuta una serie de acciones

3 Si la condición no se cumple, se ejecuta otra serie de acciones.

# **Estructuras Selectivas**
2 MULTIPLES

Se ejecutarán unas instrucciones u otras según el resultado
que se obtenga al evaluar una expresión.
Se considera que dicho resultado ha de ser de tipo ordinal,
es decir de un tipo de datos en el que cada uno de los
elementos que constituyen el tipo, excepto el primero y el
último, tiene un único predecesor y un único sucesor. Cada
grupo de instrucción se encontrará ligado con: un valor,
varios valores y separados por comas, un rango expresado
como valorInicial ... valorFinal o una mezcla de valores y
rangos.


# **Estructuras Selectivas MULTIPLES**

1 Se evalúan en orden una serie de condiciones.

2 Se ejecuta la serie de acciones asociada a la primera condición

3 que se cumpla. Si ninguna de las condiciones se cumple, se

4 ejecuta una serie de acciones alternativas.
