# **Sumar una secuencia de enteros detectando si están ordenados:**

      #include<iostream>
      using namespace std;

      int main(){
        int num, acumulador=0, num_previo=-999999;
        
        int ordenado = 1;
        
        while(num!=0){
          
          cout<<"Ingrese un numero, [si ingresa el 0(cero) termina el programa]"<<endl;
          cin>>num;
          
          acumulador = acumulador + num;
          
          if(num_previo>num){
            ordenado = 0;
          }
          
          num_previo = num;
        }
        
        cout<<"\nLa suma de los numeros ingresados es: "<<acumulador;
        
        if (ordenado == 1){
          cout<<"\nLos numeros estan ordenados"<<endl;
        }
        else{
          cout<<"\nLos numeros estan desordenados"<<endl;
        }
        
        return 0;
      }
