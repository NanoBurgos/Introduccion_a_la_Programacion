consumidores y determine qué marca fue la ganadora absoluta

de marca).

mostrar un mensaje de “Error, código inválido” y volver a pedir el
dato.

Al finalizar, el programa debe mostrar en pantalla:
“La marca preferida fue la X, con Y votos.”
